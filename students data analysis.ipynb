## Gender distributrion

```python
plt.figure(figsize=(5,6))
ax = sns.countplot(data = df,x = df['Gender'])
ax.bar_label(ax.containers[0])
plt.title('Gender distribution')
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

```python
gb
```

```python
sns.heatmap(data = gb,annot=True)
plt.title('Relationship between parents eduction and childerns score')
```

```python
# from the above charts we have concluded that the education of parents have a good impact on their scores
```

```python
gb2 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
```

## Ethnic Group Distribution

```python
groupA  = df.loc[(df['EthnicGroup'] == 'group A')].count()
groupB  = df.loc[(df['EthnicGroup'] == 'group B')].count()
groupC  = df.loc[(df['EthnicGroup'] == 'group C')].count()
groupD  = df.loc[(df['EthnicGroup'] == 'group D')].count()
groupE  = df.loc[(df['EthnicGroup'] == 'group E')].count()
l = ['group A','group B','group C','group D','group E']
mlist = [groupA['EthnicGroup'],groupB['EthnicGroup'],groupC['EthnicGroup'],groupD['EthnicGroup'],groupE['EthnicGroup']]
print(mlist)
plt.pie(mlist,labels=l,autopct="%1.2f%%")
plt.title('Ethnic Group Distribution')

```

```python
# we can also see the Ethanic group distribution using countplot
ax = sns.countplot(x=df['EthnicGroup'])
ax.bar_label(ax.containers[0])
plt.title('Ethnic Group Distribution')
```
